# Example 1 - A simple cost function

## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x ** 2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
# Plot
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

```python
def df(x):
    return 2 * x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
```

## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(new_x)  # Compute the gradient
    new_x = new_x - step_multiplier * gradient  # Update new_x

    step_size = abs(new_x - previous_x)
    x_list.append(new_x)
    slope_list.append(df(new_x))  # Append the new gradient

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
```

```python
## TASK-8 : Code the same things as they were done for the first two examples
import matplotlib.pyplot as plt
import numpy as np

# Define the new function g(x) and its derivative dg(x)
def g(x):
    return x**3 - 4*x

def dg(x):
    return 3*x**2 - 4

def gradient_descent(derivative_func, cost_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    cost_list = [cost_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        cost_list.append(cost_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list, cost_list

def cost_g(x):
    return g(x)

# Define range of x values
x_2 = np.linspace(-2, 2, 1000)

# Run gradient descent with different learning rates and track costs
initial_guess1 = 1.9
initial_guess2 = 1.5
initial_guess3 = 0.5

low_gamma = gradient_descent(derivative_func=dg, cost_func=cost_g, initial_guess=initial_guess1,
                             multiplier=0.02, precision=0.0001, max_iter=500)

mid_gamma = gradient_descent(derivative_func=dg, cost_func=cost_g, initial_guess=initial_guess2,
                             multiplier=0.001, precision=0.0001, max_iter=500)

high_gamma = gradient_descent(derivative_func=dg, cost_func=cost_g, initial_guess=initial_guess3,
                              multiplier=0.002, precision=0.0001, max_iter=500)

# Plot function, its derivative, and gradient descent paths
plt.figure(figsize=[18, 6])

# Plot for low learning rate
plt.subplot(1, 3, 1)
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.title('Low Learning Rate (0.02)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(low_gamma[1], g(np.array(low_gamma[1])), color='red', s=100, alpha=0.6)

# Plot derivative for low learning rate
plt.subplot(1, 3, 2)
plt.title('Slope of g(x)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(low_gamma[1], dg(np.array(low_gamma[1])), color='red', s=100, alpha=0.6)

# Plot cost reduction for low learning rate
plt.figure(figsize=[15, 5])
plt.title('Cost Reduction for Low Learning Rate (0.02)', fontsize=17)
plt.xlabel('Iterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(low_gamma[3], color='blue', linewidth=2)
plt.grid()

plt.show()

# Repeat for mid and high learning rates
# Plot function, its derivative, and gradient descent paths

plt.figure(figsize=[18, 6])

# Plot for mid learning rate
plt.subplot(1, 3, 1)
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.title('Mid Learning Rate (0.001)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(mid_gamma[1], g(np.array(mid_gamma[1])), color='red', s=100, alpha=0.6)

# Plot derivative for mid learning rate
plt.subplot(1, 3, 2)
plt.title('Slope of g(x)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(mid_gamma[1], dg(np.array(mid_gamma[1])), color='red', s=100, alpha=0.6)

# Plot cost reduction for mid learning rate
plt.figure(figsize=[15, 5])
plt.title('Cost Reduction for Mid Learning Rate (0.001)', fontsize=17)
plt.xlabel('Iterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(mid_gamma[3], color='green', linewidth=2)
plt.grid()

plt.show()

# Plot function, its derivative, and gradient descent paths

plt.figure(figsize=[18, 6])

# Plot for high learning rate
plt.subplot(1, 3, 1)
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.title('High Learning Rate (0.002)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(high_gamma[1], g(np.array(high_gamma[1])), color='red', s=100, alpha=0.6)

# Plot derivative for high learning rate
plt.subplot(1, 3, 2)
plt.title('Slope of g(x)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(high_gamma[1], dg(np.array(high_gamma[1])), color='red', s=100, alpha=0.6)

# Plot cost reduction for high learning rate
plt.figure(figsize=[15, 5])
plt.title('Cost Reduction for High Learning Rate (0.002)', fontsize=17)
plt.xlabel('Iterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.plot(high_gamma[3], color='red', linewidth=2)
plt.grid()

plt.show()
```

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```
